```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
mean=[0,0]
cov=[[3,2],[2,2]]
x=np.random.multivariate_normal(mean=mean,cov=cov,size=200)
```

```python
x
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],color='r')
plt.title('Scatter point of bivariate normal distribution')
plt.xlabel('X1')
plt.ylabel('X2')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x)
```

```python
x_pca
```

```python
components=pca.components_
```

```python
components
```

```python
pca.explained_variance_ratio_
```

```python
components[0],components[1]
```

```python
x
```

```python
projection_1=np.dot(x,components[0])
projection_2=np.dot(x,components[1])
projection_1
```

```python
x_pc1=projection_1*components[0][0]
x_pc2=projection_1*components[0][1]
y_pc1=projection_2*components[1][0]
y_pc2=projection_2*components[1][1]
```

```python
x_pc1
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],label='Original Data',ec='k',s=50,alpha=0.6)
plt.scatter(x_pc1,y_pc1,c='r',ec='k',marker='X',s=70,alpha=0.5,label='Projection after PC 1')
plt.scatter(x_pc2,y_pc2,c='b',ec='k',marker='X',s=70,alpha=0.5,label='Projection after PC 2')
plt.title('Linearly correlated data projected onto principal components')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
